# MATH

```python
import math

for d in dir(math):
    print(d,end="\t")
```

```python
from math import radians,degrees,pi

print(pi)
print(degrees(pi))
print(radians(90))
```

```python
from math import e,log,exp

print(pow(e,1))
print(log(e,e))
```

```python
from math import trunc,ceil,floor

x = 2.154
print(trunc(x))
print(ceil(x))
print(floor(x))
```

```python
from math import hypot,factorial

print(factorial(5))
print(hypot(2,2))
```

## Random

    #### It delivers some mechanisms allowing you to operate with pseudorandom numbers.

    #### Note the prefix pseudo - the numbers generated by the modules may look random in the sense that you cannot predict their subsequent values, but don't forget that they all are calculated using very refined algorithms

```python
import random

# this seed will control the number generation
random.seed(0)

random.random()
```

```python
print(random.randrange(0,1,1))
print(random.randint(1,10))
```

```python
sequence = [1,2,3,4]

print(random.choice(sequence))
print(random.sample(sequence,1))
```

## Platform

    The platform module lets you access the underlying platform's data, i.e., hardware, operating system, and interpreter version information.

```python
from platform import platform
print(platform()) # this will give the underlying opertaing system
```

```python
from platform import machine
print(machine()) # this will give the processor type details
```

```python
from platform import processor
print(processor())
```

```python
from platform import system
print(system())
```

```python
from platform import version
print(version())
```

```python
from platform import python_implementation,python_version_tuple

print(python_implementation())
for tp in python_version_tuple():
    print(tp)
```
