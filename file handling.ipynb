```python
# managing json file
import json

mydict = dict(a =10)

json_string = json.dumps(mydict)

with open("data//myjson.json","w") as jsf:
    jsf.write(json_string)
```

```python
# using pickling to convert python objects to string and back
import pickle 

mylist = [1,2,3,4]

mylist_string = pickle.dumps(mylist)

with open("data//mylist.pickle","wb") as myl:
    pickle.dump(mylist,myl)
    
with open("data//mylist.pickle","rb") as myl:
    ff = pickle.load(myl)
    print(ff)
```

```python
# reading data from pdf file
#!pip install PyPDF2
from PyPDF2 import PdfFileReader

with open("data/exam.pdf","rb") as f:
    pdf = PdfFileReader(f)
    print(pdf.getNumPages())
    print(pdf.getDocumentInfo())
    
    #extract data from the file
    page_obj = pdf.getPage(0)
    print(page_obj.extractText())
```
